```python
from ana_new import trans_ana
from two_step import two_step_overland
import numpy as np
from math import sqrt
from farPoint import bcdPoints


ca = trans_ana(35, 67, 100.0, 201, 0.5, 0.1, 0.01)
ca.run()

m = 5./3
alpha = 1./0.1*sqrt(0.01)
q1 = 67.0/3600/1000*100.0
ie1 = 67.0/3600/1000
S1 = m/(m+1)*alpha/ie1*(ie1*100.0/alpha)**((m+1)/m)

qBi1 = np.interp(630, ca.out_q[:, 0], ca.out_q[:, 1])/q1
SBi1 = np.interp(630, ca.out_s[:, 0], ca.out_s[:, 1])/S1

cb = two_step_overland(35, 67, 14, 100.0, 201, 0.5, 0.1, 0.01)
cb.ie1_duration(630.0)
cb.run()

SDC = bcdPoints(cb)[0]
qDi1 = SDC[-1][1]/q1
SDi1 = SDC[-1][2]/S1

da = trans_ana(16., 16.*1.91428, 100.0, 201, 0.5, 0.1, 0.01)
da.run()

ie2 = 16.*1.91428/3600./1000.
q2 = ie2*100.0
S2 = m/(m+1)*alpha/ie2*(ie2*100.0/alpha)**((m+1)/m)

d_t = np.interp(q2*qBi1, da.out_q[:, 1], da.out_q[:, 0])

db = two_step_overland(16., 16.*1.91428, 16.*1.91428/4.7857142857, 100.0, 201, 0.5, 0.1, 0.01)
db.ie1_duration(d_t)
db.run()

SDC2 = bcdPoints(db)[0]
qDi2 = SDC2[-1][1]/q2
SDi2 = SDC2[-1][2]/S2

ie3 = 83.*1.91428/3600./1000.
q3 = ie3*100.0
S3 = m/(m+1)*alpha/ie3*(ie3*100.0/alpha)**((m+1)/m)

ea = trans_ana(83., ie3*3600.*1000., 100.0, 201, 0.5, 0.1, 0.01)
ea.run()

d_t2 = np.interp(q3*qBi1, ea.out_q[:, 1], ea.out_q[:, 0])

eb = two_step_overland(83., ie3*3600.*1000., ie3*3600.*1000./4.7857142857, 100.0, 201, 0.5, 0.1, 0.01)
eb.ie1_duration(d_t2)
eb.run()

SDC3 = bcdPoints(eb)[0]
qDi3 = SDC3[-1][1]/q3
SDi3 = SDC3[-1][2]/S3
```

```python
import matplotlib.pyplot as plt


SDC = np.array(SDC)
SDC2 = np.array(SDC2)
SDC3 = np.array(SDC3)
SDC[:, 1] = SDC[:, 1]/q1
SDC[:, 2] = SDC[:, 2]/S1
SDC2[:, 1] = SDC2[:, 1]/q2
SDC2[:, 2] = SDC2[:, 2]/S2
SDC3[:, 1] = SDC3[:, 1]/q3
SDC3[:, 2] = SDC3[:, 2]/S3

titleString = '$q_B/q_{i_2}$ = %6.3f $s_B/s_{i_2}$ = %6.3f \n $q_D/q_{i_2}$ = %6.3f $s_D/s_{i_2}$ = %6.3f' % (qBi1, SBi1, qDi1, SDi1)

plt.plot(SDC[:, 1], SDC[:, 2], 'r-', lw=3.0, label='35-67-14')
plt.plot(SDC3[:, 1], SDC3[:, 2], color='beige', lw=2.0, label='83.0-158.88-34.0')
plt.plot(SDC2[:, 1], SDC2[:, 2], 'c--', label='16-30.63-6.40')

plt.xlabel('$\dot{q}$')
plt.ylabel('$\dot{s}$')
plt.title(titleString)
plt.legend(loc='upper left')
plt.savefig('SDC compare.png')
```
