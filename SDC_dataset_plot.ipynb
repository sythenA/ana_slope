```python
import farPoint as fp

a = fp.SDRCurves(100.0, 201, 0.5, 0.1, 0.01, 2, 4, useFile='ks1_SDRrec.pick')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


qBDRatio = a.qBDRatio
sBDRatio = a.sBDRatio
endSlope = a.endSlope

qBDRatio = np.array(qBDRatio)
sBDRatio = np.array(sBDRatio)
endSlope = np.array(endSlope)

qBDRatio = (qBDRatio - min(qBDRatio))/(max(qBDRatio) - min(qBDRatio))
sBDRatio = (sBDRatio - min(sBDRatio))/(max(sBDRatio) - min(sBDRatio))
endSlope = endSlope/max(endSlope)


fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(qBDRatio, sBDRatio, endSlope, '.')
# ax.set_zbound(lower=-0.1, upper=0.1)
# ax.set_xlim([0.0, 0.075])
# ax.set_ylim([0.0, 0.075])

ax.set_xlabel('$q_B/q_D$ Normalized')
ax.set_ylabel('$S_B/S_D$ Normalized')
ax.set_zlabel('$dq/dS$ Normalized')
ax.set_title('Secondary Drying Curve DataSet(Normalized and Focus)')
plt.savefig('SDC Data Distribution.png', dpi=150)
```
