```python
import steady_fit as st

std = st.fit_steady(save_name='ks1', folder='Steady_fit')
std.param(100.0, 0.1, 0.5, 0.01)
std.steady_SQ()
std.load()
```

```python
import bezeir_fit as bf
import numpy as np

# Start from zero
std.steady_q.tolist().insert(0, 0.)
std.steady_s.tolist().insert(0, 0.)

t = np.arange(0., 1.0001, 1./10)
Q = np.interp(t, std.ie/max(std.ie), std.steady_q)
S = np.interp(t, std.ie/max(std.ie), std.steady_s)

std_SQ = np.array([zip(Q, S)])[0]
p,t = bf.fit(std_SQ, 2, 1.0E-8)
print p

p[0] = [0, 0]
p[-1] = std_SQ[-1]
t = np.arange(0., 1.0001, 1./100)
f_std_SQ = bf.gen(t, 2, p)
```

```python
import matplotlib.pyplot as plt

plt.plot(std_SQ[:, 0], std_SQ[:, 1], ls='--')
plt.plot(f_std_SQ[:, 0], f_std_SQ[:, 1])
plt.show()
```
